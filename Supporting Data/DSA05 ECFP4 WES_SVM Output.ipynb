```python

# Importing packages 
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem, PandasTools

import numpy as np
from numpy import sqrt, argmax

import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
from pandas import DataFrame

import os

import xgboost as xgb
import lightgbm as lgb

from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.metrics import accuracy_score, cohen_kappa_score, matthews_corrcoef, roc_curve, precision_recall_curve, roc_auc_score, make_scorer
from sklearn.metrics import f1_score, balanced_accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.feature_selection import VarianceThreshold, RFE
from sklearn.calibration import calibration_curve

import joblib
from sklearn.model_selection import cross_val_score
from imblearn.metrics import geometric_mean_score

import optuna

# Applicability Domain Functions
from sklearn.metrics.pairwise import euclidean_distances

def calculate_applicability_domain(X_train, X_test, threshold=0.5):
    distances = euclidean_distances(X_train, X_train)
    avg_distance = np.mean(distances)
    std_distance = np.std(distances)
    
    # Define applicability domain threshold
    apd_threshold = avg_distance + threshold * std_distance
    
    # Calculate test distances to training data
    test_distances = euclidean_distances(X_test, X_train)
    min_distances = np.min(test_distances, axis=1)
    
    # Calculate applicability scores
    applicability_scores = 1 - (min_distances / apd_threshold)
    
    # Ensure the applicability scores are between 0 and 1
    applicability_scores = np.clip(applicability_scores, 0, 1)
    
    # Convert to percentage
    applicability_scores_percentage = applicability_scores * 100
    
    return applicability_scores_percentage

def calculate_coverage(applicability_scores_fold, threshold=0.5):
    within_domain_count = np.sum(applicability_scores_fold >= threshold * 100)
    total_data_points = len(applicability_scores_fold)
    coverage = (within_domain_count / total_data_points) * 100.0
    return coverage

# Reading molecules and activity (0 and 1) from SDF
fname = "../data/dataset_ss_DSA05_WES_GHS_BIN_Binary.sdf"
df0 = PandasTools.LoadSDF(fname, smilesName='SMILES2', includeFingerprints=False)
df0["is_active"] = df0["is_active"]

df0.head()
target_count = df0.is_active.value_counts()
print('Class 0:', target_count[1])
print('Class 1:', target_count[0])
print('Proportion:', round(target_count[0] / target_count[1], 2), ': 1')

target_count.plot(kind='bar', title='Count (target)')

# Fingerprint calculation
fname = "../data/dataset_ss_DSA05_WES_GHS_BIN_Binary.sdf"

mols = []
y = []
for mol in Chem.SDMolSupplier(fname):
    if mol is not None:
        mols.append(mol)
        y.append(mol.GetIntProp("is_active")) # target column

# Generate binary morgan fingerprint with radius 2
fp = [AllChem.GetMorganFingerprintAsBitVect(m, 2, 2048, useFeatures=False) for m in mols]

def rdkit_numpy_convert(fp):
    output = []
    for f in fp:
        arr = np.zeros((1,))
        DataStructs.ConvertToNumpyArray(f, arr)
        output.append(arr)
    return np.asarray(output)

x = rdkit_numpy_convert(fp)

# Filter Low Variance Descriptors
def filter_low_variance_features(X, threshold=0.01):
    selector = VarianceThreshold(threshold)
    X_reduced = selector.fit_transform(X)
    return X_reduced, selector.get_support(indices=True)

X_reduced, selected_features_low_variance = filter_low_variance_features(x)
x = X_reduced

# Feature Selection Using Recursive Feature Elimination (RFE)
def select_features_rfe(X, y, model, n_features_to_select=50):
    rfe = RFE(estimator=model, n_features_to_select=n_features_to_select)
    X_reduced = rfe.fit_transform(X, y)
    return X_reduced, rfe.support_

# Using RandomForestClassifier for feature selection
rf_model = RandomForestClassifier()
X_selected, selected_features_rfe = select_features_rfe(x, y, rf_model)
x = X_selected

# Save selected feature indices
selected_features = selected_features_low_variance[selected_features_rfe]

# Specify the folder where you want to save the selected features
save_folder = "../models"

# Save the selected features indices to the specified folder
save_path = os.path.join(save_folder, 'selected_features_svm_binary.npy')
np.save(save_path, selected_features)

# Check the number of compounds
len(x)

# Check whether the data set is balanced
sum(y) / len(y)

## Data splitting
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

## RF Model building - Bayesian hyperparameter search
def objective_rf(trial):
    n_estimators = trial.suggest_int('n_estimators', 100, 1000)
    max_depth = trial.suggest_int('max_depth', 2, 100)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 20)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 20)

    clf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, min_samples_leaf=min_samples_leaf, min_samples_split=min_samples_split, random_state=42)
    return cross_val_score(clf, x, y, n_jobs=-1, cv=cv, scoring=make_scorer(geometric_mean_score)).mean()

study_rf = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler(seed=42))
study_rf.optimize(objective_rf, n_trials=10)
best_params_rf = study_rf.best_params

probs_classes_rf = []
y_test_all_rf = []
applicability_scores_all_rf = []

for train_index, test_index in cv.split(x, y):
    rf_clf = RandomForestClassifier(**best_params_rf)
    X_train_folds = x[train_index]
    y_train_folds = np.array(y)[train_index.astype(int)]
    X_test_fold = x[test_index]
    y_test_fold = np.array(y)[test_index.astype(int)]
    
    rf_clf.fit(X_train_folds, y_train_folds)
    y_pred = rf_clf.predict_proba(X_test_fold)
    probs_classes_rf.append(y_pred)
    y_test_all_rf.append(y_test_fold)
    
    applicability_scores_fold = calculate_applicability_domain(X_train_folds, X_test_fold)
    applicability_scores_all_rf.append(applicability_scores_fold)

coverages_rf = [calculate_coverage(applicability_scores_fold) for applicability_scores_fold in applicability_scores_all_rf]
average_coverage_rf = np.mean(coverages_rf)
print(f"Average coverage for Random Forest: {average_coverage_rf}")

# Check performance of each fold
fold_preds_rf = [(y_pred[:, 1] > 0.5).astype(int) for y_pred in probs_classes_rf]
fold_exps_rf = y_test_all_rf

for i in range(len(fold_preds_rf)):
    bacc = metrics.balanced_accuracy_score(fold_exps_rf[i], fold_preds_rf[i])
    print(f"Balanced accuracy (fold {i+1}) = ", bacc)

# Check mean performance of folds
probs_classes_rf = np.concatenate(probs_classes_rf)
y_experimental_rf = np.concatenate(y_test_all_rf)
pred_rf = (probs_classes_rf[:, 1] > 0.5).astype(int)

def calc_statistics(y, pred):
    confusion = confusion_matrix(y, pred)
    TP = confusion[1, 1]
    TN = confusion[0, 0]
    FP = confusion[0, 1]
    FN = confusion[1, 0]
    #classification_error = round(1 - accuracy_score(y, pred), 2)
    sensitivity = round(recall_score(y, pred), 2)
    specificity = round(TN / (TN + FP), 2)
    ccr = round((sensitivity + specificity) / 2, 2)
   # false_positive_rate = round(FP / float(TN + FP), 2)
    #false_negative_rate = round(FN / float(TP + FN), 2)
    positive_pred_value = round(TP / float(TP + FP), 2)
    negative_pred_value = round(TN / float(TN + FN), 2)
    auc = round(roc_auc_score(y, pred), 2)  # Fix: AUC should be calculated with actual and predicted probabilities
    bacc = round(balanced_accuracy_score(y, pred), 2)
   # f1 = round(f1_score(y, pred), 2)
    #accuracy = round(accuracy_score(y, pred), 2)
    #mcc = round(matthews_corrcoef(y, pred), 2)
    #kappa = round(cohen_kappa_score(y, pred), 2)

    statistics = pd.DataFrame({'Bal-acc': bacc, "Sensitivity": sensitivity, 'CCR': ccr, "Specificity": specificity, "PPV": positive_pred_value, 
                               "NPV": negative_pred_value, 
                               #'Kappa': kappa, 'AUC': auc, 'MCC': mcc, 'Accuracy': accuracy, 
                               #"Classification error": classification_error, "False positive rate": false_positive_rate, 
                               #"False negative rate": false_negative_rate, 'F1-score': f1
                               }, 
                               index=[0])
    return statistics

statistics_rf = calc_statistics(y_experimental_rf, pred_rf)
auc_rf = round(roc_auc_score(y_experimental_rf, probs_classes_rf[:, 1]), 2)  # Fix: AUC with actual and predicted probabilities
statistics_rf['AUC'] = auc_rf
model_type = "morgan_r2_2048_rf"
result_type = "uncalibrated"
metrics_rf_uncalibrated = statistics_rf
metrics_rf_uncalibrated['model'] = model_type
metrics_rf_uncalibrated['result_type'] = result_type
metrics_rf_uncalibrated['Coverage'] = average_coverage_rf
print("Uncalibrated model:")
print(metrics_rf_uncalibrated)

# Calibrate RF Model
probs = probs_classes_rf[:, 1]
fop, mpv = calibration_curve(y_experimental_rf, probs, n_bins=10)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(mpv, fop, marker='.')
#plt.savefig('../results/calibration_model_morgan_r2_2048_rf_dataset-balanced_std.png', dpi=300)

# Use ROC-Curve and Gmean to select a threshold for calibration
yhat = probs_classes_rf[:, 1]
fpr, tpr, thresholds = roc_curve(y_experimental_rf, yhat)
gmeans = sqrt(tpr * (1-fpr))
ix = argmax(gmeans)
print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='RF')
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
#plt.savefig('../results/calibration_curve_morgan_r2_2048_rf_dataset-balanced_std.png', dpi=300)
threshold_roc = thresholds[ix]

# Optimal Threshold for Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_experimental_rf, yhat)
fscore = (2 * precision * recall) / (precision + recall)
ix = argmax(fscore)
print('Best Threshold=%f, F-Score=%.3f' % (thresholds[ix], fscore[ix]))
threshold_prc = thresholds[ix]

# Calibrated RF Model Statistics
pred_rf_calibrated = (probs_classes_rf[:, 1] > threshold_roc).astype(int)
statistics_rf_calibrated = calc_statistics(y_experimental_rf, pred_rf_calibrated)
auc_rf_calibrated = round(roc_auc_score(y_experimental_rf, probs_classes_rf[:, 1]), 2)  # AUC remains the same
statistics_rf_calibrated['AUC'] = auc_rf_calibrated
result_type = "calibrated"
metrics_rf_calibrated = statistics_rf_calibrated
metrics_rf_calibrated['model'] = model_type
metrics_rf_calibrated['result_type'] = result_type
metrics_rf_calibrated['Coverage'] = average_coverage_rf
metrics_rf_calibrated['calibration_threshold'] = threshold_roc
print("Calibrated model:")
print(metrics_rf_calibrated)

# SVM Model building - Bayesian hyperparameter search
def objective_svm(trial):
    C = trial.suggest_loguniform('C', 1e-6, 1e6)
    gamma = trial.suggest_loguniform('gamma', 1e-6, 1e1)
    kernel = trial.suggest_categorical('kernel', ['rbf'])

    svc = SVC(probability=True, C=C, gamma=gamma, kernel=kernel)
    score = cross_val_score(svc, x, y, cv=cv, scoring=make_scorer(geometric_mean_score), n_jobs=-1)
    return score.mean()

study_svm = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler(seed=42))
study_svm.optimize(objective_svm, n_trials=50)
best_params_svm = study_svm.best_params

probs_classes_svm = []
y_test_all_svm = []
applicability_scores_all_svm = []

for train_index, test_index in cv.split(x, y):
    svm_clf = SVC(**best_params_svm, probability=True)
    X_train_folds = x[train_index]
    y_train_folds = np.array(y)[train_index.astype(int)]
    X_test_fold = x[test_index]
    y_test_fold = np.array(y)[test_index.astype(int)]
    
    svm_clf.fit(X_train_folds, y_train_folds)
    y_pred = svm_clf.predict_proba(X_test_fold)
    probs_classes_svm.append(y_pred)
    y_test_all_svm.append(y_test_fold)
    
    applicability_scores_fold = calculate_applicability_domain(X_train_folds, X_test_fold)
    applicability_scores_all_svm.append(applicability_scores_fold)

coverages_svm = [calculate_coverage(applicability_scores_fold) for applicability_scores_fold in applicability_scores_all_svm]
average_coverage_svm = np.mean(coverages_svm)
print(f"Average coverage for SVM: {average_coverage_svm}")

# Check performance of each fold
fold_preds_svm = [(y_pred[:, 1] > 0.5).astype(int) for y_pred in probs_classes_svm]
fold_exps_svm = y_test_all_svm

for i in range(len(fold_preds_svm)):
    bacc = metrics.balanced_accuracy_score(fold_exps_svm[i], fold_preds_svm[i])
    print(f"Balanced accuracy (fold {i+1}) = ", bacc)

# Check mean performance of folds
probs_classes_svm = np.concatenate(probs_classes_svm)
y_experimental_svm = np.concatenate(y_test_all_svm)
pred_svm = (probs_classes_svm[:, 1] > 0.5).astype(int)

statistics_svm = calc_statistics(y_experimental_svm, pred_svm)
auc_svm = round(roc_auc_score(y_experimental_svm, probs_classes_svm[:, 1]), 2)  # Fix: AUC with actual and predicted probabilities
statistics_svm['AUC'] = auc_svm
model_type = "morgan_r2_2048_svm"
result_type = "uncalibrated"
metrics_svm_uncalibrated = statistics_svm
metrics_svm_uncalibrated['model'] = model_type
metrics_svm_uncalibrated['result_type'] = result_type
metrics_svm_uncalibrated['Coverage'] = average_coverage_svm
print("Uncalibrated model:")
print(metrics_svm_uncalibrated)

# Calibrate SVM Model
probs = probs_classes_svm[:, 1]
fop, mpv = calibration_curve(y_experimental_svm, probs, n_bins=10)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(mpv, fop, marker='.')
#plt.savefig('../results/calibration_model_morgan_r2_2048_svm_dataset-balanced_std.png', dpi=300)

# Use ROC-Curve and Gmean to select a threshold for calibration
yhat = probs_classes_svm[:, 1]
fpr, tpr, thresholds = roc_curve(y_experimental_svm, yhat)
gmeans = sqrt(tpr * (1-fpr))
ix = argmax(gmeans)
print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='SVM')
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
#plt.savefig('../results/calibration_curve_morgan_r2_2048_svm_dataset-balanced_std.png', dpi=300)
threshold_roc = thresholds[ix]

# Optimal Threshold for Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_experimental_svm, yhat)
fscore = (2 * precision * recall) / (precision + recall)
ix = argmax(fscore)
print('Best Threshold=%f, F-Score=%.3f' % (thresholds[ix], fscore[ix]))
threshold_prc = thresholds[ix]

# Calibrated SVM Model Statistics
pred_svm_calibrated = (probs_classes_svm[:, 1] > threshold_roc).astype(int)
statistics_svm_calibrated = calc_statistics(y_experimental_svm, pred_svm_calibrated)
auc_svm_calibrated = round(roc_auc_score(y_experimental_svm, probs_classes_svm[:, 1]), 2)  # AUC remains the same
statistics_svm_calibrated['AUC'] = auc_svm_calibrated
result_type = "calibrated"
metrics_svm_calibrated = statistics_svm_calibrated
metrics_svm_calibrated['model'] = model_type
metrics_svm_calibrated['result_type'] = result_type
metrics_svm_calibrated['Coverage'] = average_coverage_svm
metrics_svm_calibrated['calibration_threshold'] = threshold_roc
print("Calibrated model:")
print(metrics_svm_calibrated)

import joblib

# Save the calibrated SVM model and the threshold
calibrated_model_data = {
    'model': svm_clf,
    'threshold': threshold_roc
}

joblib.dump(calibrated_model_data, "../models/ss_DSA05_binary_morgan_r2_2048_svm_calibrated_with_threshold.joblib")

joblib.dump(x, "../models/DSA05_BINARY_SVM_x_train.pkl")

## LightGBM Model building - Bayesian hyperparameter search
def objective_lgb(trial):
    param = {
        'objective': 'binary',
        'metric': 'binary_logloss',
        'verbosity': -1,
        'boosting_type': 'gbdt',
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000),
        'max_depth': trial.suggest_int('max_depth', 2, 100),
        'num_leaves': trial.suggest_int('num_leaves', 2, 256),
        'learning_rate': trial.suggest_loguniform('learning_rate', 1e-8, 1.0),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
        'subsample': trial.suggest_uniform('subsample', 0.4, 1.0),
        'colsample_bytree': trial.suggest_uniform('colsample_bytree', 0.4, 1.0)
    }

    gbm = lgb.LGBMClassifier(**param)
    return cross_val_score(gbm, x, y, n_jobs=-1, cv=cv, scoring=make_scorer(geometric_mean_score)).mean()

study_lgb = optuna.create_study(direction='maximize', sampler=optuna.samplers.TPESampler(seed=42))
study_lgb.optimize(objective_lgb, n_trials=10)
best_params_lgb = study_lgb.best_params

probs_classes_lgb = []
y_test_all_lgb = []
applicability_scores_all_lgb = []

for train_index, test_index in cv.split(x, y):
    gbm_clf = lgb.LGBMClassifier(**best_params_lgb)
    X_train_folds = x[train_index]
    y_train_folds = np.array(y)[train_index.astype(int)]
    X_test_fold = x[test_index]
    y_test_fold = np.array(y)[test_index.astype(int)]
    
    gbm_clf.fit(X_train_folds, y_train_folds)
    y_pred = gbm_clf.predict_proba(X_test_fold)
    probs_classes_lgb.append(y_pred)
    y_test_all_lgb.append(y_test_fold)
    
    applicability_scores_fold = calculate_applicability_domain(X_train_folds, X_test_fold)
    applicability_scores_all_lgb.append(applicability_scores_fold)

coverages_lgb = [calculate_coverage(applicability_scores_fold) for applicability_scores_fold in applicability_scores_all_lgb]
average_coverage_lgb = np.mean(coverages_lgb)
print(f"Average coverage for LightGBM: {average_coverage_lgb}")

# Check performance of each fold
fold_preds_lgb = [(y_pred[:, 1] > 0.5).astype(int) for y_pred in probs_classes_lgb]
fold_exps_lgb = y_test_all_lgb

for i in range(len(fold_preds_lgb)):
    bacc = metrics.balanced_accuracy_score(fold_exps_lgb[i], fold_preds_lgb[i])
    print(f"Balanced accuracy (fold {i+1}) = ", bacc)

# Check mean performance of folds
probs_classes_lgb = np.concatenate(probs_classes_lgb)
y_experimental_lgb = np.concatenate(y_test_all_lgb)
pred_lgb = (probs_classes_lgb[:, 1] > 0.5).astype(int)

statistics_lgb = calc_statistics(y_experimental_lgb, pred_lgb)
auc_lgb = round(roc_auc_score(y_experimental_lgb, probs_classes_lgb[:, 1]), 2)  # Fix: AUC with actual and predicted probabilities
statistics_lgb['AUC'] = auc_lgb
model_type = "morgan_r2_2048_lgb"
result_type = "uncalibrated"
metrics_lgb_uncalibrated = statistics_lgb
metrics_lgb_uncalibrated['model'] = model_type
metrics_lgb_uncalibrated['result_type'] = result_type
metrics_lgb_uncalibrated['Coverage'] = average_coverage_lgb
print("Uncalibrated model:")
print(metrics_lgb_uncalibrated)

# Calibrate LightGBM Model
probs = probs_classes_lgb[:, 1]
fop, mpv = calibration_curve(y_experimental_lgb, probs, n_bins=10)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(mpv, fop, marker='.')
#plt.savefig('../results/calibration_model_morgan_r2_2048_lgb_dataset-balanced_std.png', dpi=300)

# Use ROC-Curve and Gmean to select a threshold for calibration
yhat = probs_classes_lgb[:, 1]
fpr, tpr, thresholds = roc_curve(y_experimental_lgb, yhat)
gmeans = sqrt(tpr * (1-fpr))
ix = argmax(gmeans)
print('Best Threshold=%f, G-Mean=%.3f' % (thresholds[ix], gmeans[ix]))
plt.plot([0,1], [0,1], linestyle='--', label='No Skill')
plt.plot(fpr, tpr, marker='.', label='LightGBM')
plt.scatter(fpr[ix], tpr[ix], marker='o', color='black', label='Best')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
#plt.savefig('../results/calibration_curve_morgan_r2_2048_lgb_dataset-balanced_std.png', dpi=300)
threshold_roc = thresholds[ix]

# Optimal Threshold for Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_experimental_lgb, yhat)
fscore = (2 * precision * recall) / (precision + recall)
ix = argmax(fscore)
print('Best Threshold=%f, F-Score=%.3f' % (thresholds[ix], fscore[ix]))
threshold_prc = thresholds[ix]

# Calibrated LightGBM Model Statistics
pred_lgb_calibrated = (probs_classes_lgb[:, 1] > threshold_roc).astype(int)
statistics_lgb_calibrated = calc_statistics(y_experimental_lgb, pred_lgb_calibrated)
auc_lgb_calibrated = round(roc_auc_score(y_experimental_lgb, probs_classes_lgb[:, 1]), 2)  # AUC remains the same
statistics_lgb_calibrated['AUC'] = auc_lgb_calibrated
result_type = "calibrated"
metrics_lgb_calibrated = statistics_lgb_calibrated
metrics_lgb_calibrated['model'] = model_type
metrics_lgb_calibrated['result_type'] = result_type
metrics_lgb_calibrated['Coverage'] = average_coverage_lgb
metrics_lgb_calibrated['calibration_threshold'] = threshold_roc
print("Calibrated model:")
print(metrics_lgb_calibrated)

## Combine all results into a single DataFrame
frames = [metrics_rf_uncalibrated, metrics_svm_uncalibrated, metrics_lgb_uncalibrated, 
          metrics_rf_calibrated, metrics_svm_calibrated, metrics_lgb_calibrated]

result = pd.concat(frames)

column_names = ["model", "CCR", "Sensitivity", "Specificity", "PPV", "NPV", "AUC",  
                "Coverage", "result_type", "calibration_threshold"]

result = result.reindex(columns=column_names)

result

# Save the results to an Excel file - CHANGE EVERY TIME
result.to_excel("../models/ss_DSA05_binary_morgan_r2_2048_WES.xlsx", sheet_name= "WES")
```
